For starters, the null hypothesis is made in relation to the population. Any tests performed to either confirm or reject this hypothesis are done "thinking" that the main results point to the actual population. The population parameter refers to a real statistical value from a real population, such as the mean (average) of the height of a UofT student. In essence, we use sample statistics to provide a prediction or estimate of the actual population parameter, which in turn helps us reject or confirm the null hypothesis.

The reason why a smaller p-value makes the null hypothesis become 'more ridiculous' is pretty simple. The p-value is the probability of finding a value equal to or more extreme (that is, 'bigger') than our observed statistic. The observed statistic is obtained from our sample. Having said this, when we assume the null hypothesis is true, we assume that finding values close to our observed statistic would not be too uncommon. This would mean that in a world where the null hypothesis is true, our sample would make sense. If the p-value is small, it means that in this world, encountering values such as our observed statistic is very rare. This would imply that our sample does not make sense. This, in turn, gives us enough evidence to reject that the null hypothesis is true. Note that we cannot say the null hypothesis is true, we can only say there is evidence that suggests it can be true or there is not enough evidence to reject it. In any case, we say the smaller the p-value, the more 'ridiculous' the world where the null hypothesis is true because it means our sample and observed statistic would make no sense.

### Problem introduction

The original experiment carried out by Fisher had as its purpose to test whether or not Bristol had the ability to properly distinguish between milk being poured before and after the tea. The question was whether there was any statistical evidence suggesting that her guesses were not just products of chance. In the case of STA130 students, the scope of the experiment is perhaps a bit more limited. Due to the size of the sample and the nature of the experiment, a single student's ability to discern tea and milk being poured in a particular order cannot really be studied in depth. The scope of this experiment is to observe, on average, whether STA130 students are good at discerning the order in which tea and milk are poured or not, regardless of whether they are just lucky or are have a real ability to do so. This is definitely a much more abstract observation than Fisher's original experiment. 

The null hypothesis for this experiment is to assume that STA130 students are not, on average, good at discerning the difference in milk pouring order. That is, they can't pick the correct state of the tea. In order to test this hypothesis, the data will be bootstrapped assuming the null hypothesis is true.

$H_0$: On average, STA130 students can't correctly choose the state of the tea.
$H_A$: On average, STA130 can correctly choose the state of the tea.

In order to test assuming the null hypothesis:
$H_0$: There is a 50% chance that a STA130 student chooses the incorrect state of the tea.

For this experiment, the $\alpha$ significance is set at 0.05.

### Quantitative Analysis and Methodology

This first piece of code initializes the sample and the observed statistic.

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Creating the sample. 49 students have True values. 31 have False values.
data = np.array([1]*49 + [0]*31)

# Observed statistic of the sample (the mean)
observed_statistic = np.mean(data)
print(f'Observed Statistic: {observed_statistic}')
```

This next piece of code bootstraps the data and compares the results to the null hypothesis proportion (50%), in order to get the 'as or more extreme' values. In this case, these values are shown in a symmetrical manner, meaning that these extreme values appear on both sides of the distribution of the bootstrapped data.

```python
# This is the proportion of the null hypothesis (50% percent)
null_proportion = 0.5

np.random.seed(1) # Making it reproducible
num_simulations = 10000
n_size = len(data)

sampled_statistics = np.zeros(num_simulations) 
# Empty array to store the sampled statistics from the bootstrapped data.

for i in range(num_simulations): # Perform simulations.
    
    random_tea_choosing = np.random.choice([0,1], size=n_size, replace=True) # Boostrapping with replacement.
    sampled_statistics[i] = random_tea_choosing.mean()

    
simulated_statistics = sampled_statistics

# Calculate 'as or more extreme' values symmetrically
extreme_values = \
    abs(simulated_statistics - null_proportion) >= \
    abs(observed_statistic - null_proportion)

# Calculate p-value
p_value = extreme_values.sum() / num_simulations
```

The last bit of code is just a visual (or an attempt of) representation of what the bootstrapped data looks like and where the extreme values lie.

```python
# Show results
print("Number of Simulations: ", num_simulations)
print("Number of simulated statistics (under H₀)\nthat are 'as or more extreme' than the observed statistic: ", extreme_values.sum())
print("p-value (simulations 'as or more extreme' / total simulations): ", p_value)

# Plotting the histogram of simulated statistics
fig = px.histogram(simulated_statistics, nbins=50, title='Simulated Distribution of Proportion Differences')

# Add vertical lines for the observed statistic and its mirror (symmetric extremes)
fig.add_vline(x=observed_statistic, line_width=3, line_dash="dash", line_color="red", 
              annotation_text="Observed Statistic", annotation_position="top right")

# Add symmetric line on the opposite side
symmetric_extreme = 2 * null_proportion - observed_statistic
fig.add_vline(x=symmetric_extreme, line_width=3, line_dash="dash", line_color="blue", 
              annotation_text="Symmetric Extreme", annotation_position="bottom right")

# Highlight the "as or more extreme" regions
fig.add_vrect(x0=-np.inf, x1=symmetric_extreme, fillcolor="blue", opacity=0.2, line_width=0)
fig.add_vrect(x0=observed_statistic, x1=np.inf, fillcolor="red", opacity=0.2, line_width=0)

# Update the layout to expand the x-axis
fig.update_layout(
    xaxis_title='Proportion of True Values (Simulated)',  # Add axis label
    yaxis_title='Frequency',
    xaxis_range=[simulated_statistics.min() - 0.05, simulated_statistics.max() + 0.05],  # Expand axis limits
    width=900,  # Increase plot width to give more horizontal space
    height=500  # Adjust the height as needed
)

# Show the plot
fig.show(renderer = 'png')
```

### Findings and Discussion

Since the p-value is smaller than the previously established $\alpha$ significance level, the conclusion of the experiment is that the null hypothesis can be comfortably rejected in favor of assuming that STA130 can indeed, on average, correctly choose the state of the tea.

### 1. **Sample Data**
   - We started with a sample consisting of 49 `True` values (1s) and 31 `False` values (0s).
   - Our goal is to perform formal hypothesis testing using bootstrapping to simulate the null hypothesis and calculate a p-value for the observed statistic.

### 2. **Observed Statistic**
   - The observed statistic is calculated as the proportion of `True` values in the sample.
   - Formula: `observed_statistic = (data == 1).mean()`, which gives a value of **0.6125**.

### 3. **Null Hypothesis Setup**
   - The null hypothesis assumes that the true proportion is 0.5 (i.e., 50% `True` values).
   - We perform 10,000 simulations where, in each simulation, we bootstrap a sample of the same size as the original data, randomly selecting values with replacement.
   - The simulated statistic for each bootstrapped sample is stored.

### 4. **Extreme Values & p-value Calculation**
   - For each simulation, we check if the simulated statistic is "as or more extreme" than the observed statistic, symmetrically, with respect to the null proportion.
   - The p-value is computed as the proportion of simulated statistics that are as or more extreme than the observed statistic.

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one). 
    
> Similarly, analagously, a "significant result" used to reject the null hypothsis could alternatively be a Type I error (which means a test actually incorrectly rejected a null hypothesis when it was actually true)... we're only providing a measure of evidence against the null hypothesis... but the evidence could still incorrectly suggest the wrong conclusion... it really depends on how strong the evidence is...
>
> - all of which is why just interpreting p-values using the table above is a good idea...

3. The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. For example, distributional assumptions about the population, estimated "plug-in" values that can used to simplify the problem calculations, and the reliance upon "random sampling", etc...<br><br>
           
4. Drawing overly broad conclusions, or making recommendations based on findings that reject the null hypothesis in a specific context is fraught with the problematic risks of overgeneralization errors. Further exacerbating this issue, null hypotheses are typically so called "point null hypotheses" which is meant to emphasize that they are mathematically increadibly sharply specific; whereas, alternative hypotheses are usually very unspecific. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
    
As an example really giving a demonstrating this, consider rejecting a null hypothesis that there is no correlation between rain and pizza's delivered. Such a decision doesn't specify what the actual hypothetical correlation might be. In fact, it doesn't even indicate if there are more or less pizzas delivered when it rains... 

> which, actually, shows very clearly why statistical inference using hypothesis testing is inferior to statistical inference based on confidence intervals...
> 
> - a confidence interval provides a range of plausible values of what the parameter in question might be; whereas, ...
> - trying to more clearly address what the plausible values of the parameter in question might be on the basis of hypothesis testing would require conducting further experiements to continously reject increasingly detailed hypothesies to narrow down what the alternative hypothesis might actually include... which would indeed be an utterly vapid misuse of the intended purpose of hypothesis testing entrprise...
