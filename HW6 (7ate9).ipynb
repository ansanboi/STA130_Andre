The theoretical simple linear regression model is, quite obviously, a linear equation. It represents the relationship between the independent predictor variable $x_i$ and the outcome variable $Y$. $\beta_0$ represents the intercept (when the predictor $x_i$ is 0), $\beta_1$ is the slope and $\epsilon_i$ is the error term, which is normally distributed. The error term is what makes each value taken by the outcome variable to be normally distributed. It shows the 'error' or deviation of the outcome variable from the linear relationship (the linear equation of the linear regression model).

The theoretical linear regression model is what represents the 'real' linear relationship in the population.

````python
import numpy as np
import plotly.graph_objs as go
from scipy.stats import norm

# Parameters for the theoretical model
np.random.seed(42)  # for reproducibility
n = 100  # number of data points
beta_0 = 5  # intercept
beta_1 = 2  # slope
sigma = 1.5  # standard deviation of the error term

# Simulate independent variable x_i
x_i = np.random.uniform(0, 10, n)

# Generate normally distributed errors
epsilon = np.random.normal(0, sigma, n)

# Simulate dependent variable Y based on the linear model
Y = beta_0 + beta_1 * x_i + epsilon

# Visualize the data
scatter_trace = go.Scatter(x=x_i, y=Y, mode='markers', name='Simulated Data')
line_trace = go.Scatter(x=x_i, y=beta_0 + beta_1 * x_i, mode='lines', name='Theoretical Line')

# Layout for the plot
layout = go.Layout(
    title='Simple Linear Regression (Simulated Data)',
    xaxis=dict(title='x_i'),
    yaxis=dict(title='Y'),
    showlegend=True
)

# Create figure and plot
fig = go.Figure(data=[scatter_trace, line_trace], layout=layout)
fig.show(renderer = 'png')
````

The points are just a simulation of the outcome variable $\text Y$ when multiple $\text x_i$ values are given (by adding the error terms to each point, which is normally distributed). All these randomly sampled points from the theoretical equation give the first scatter plot accompanied by the theoretical line (the random part comes from the error term that is sampled from a normal distribution, since the coefficients are fixed). This theoretical line represents the (theoretical) linear relationship of the population (the error term is introduced when sampling $\text Y$ values).

The second line is built from a regression model that finds the best fit line for the sampled $\text Y$ values, meaning that this is the best estimate of the simulated data yielded by the theoretical linear equation (or relationship). By resampling from the theoretical linear regression, one could get different best-fit models (using ols). This is because of the error term $\epsilon_i$ that is added to each simulated value of $\text Y$. This best-fit model has a very low chance of ever being exaclty the same as the original theoretical linear relationship, since it will always find the line that has the lowest distance to the scattered points, while the data yielded by the introduction of the error term (the error term introduces the distance of the scattered points to the line) is arbitrarily 'far away' from the line.

The fitted line is built from an OLS (Ordinary Least Squares) regression model. What this model does is estimate the coefficients of the linear relationship (the equation) by finding the minimum sum of the squared residuals. The residuals are the difference between the observed data points and the predicted points (points along the theoretical line). This is the same as calculating the distance of each data point to the theoretical line.

By minimizing this sum, the model is able to find the best-fit line (that line from which all the data points have the least distance to the theoretical line).

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Define parameters
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': Y})  # Lowercase 'y' here

# OLS Model (correct the formula to match the column name in df)
model_data_specification = smf.ols("y ~ x", data=df)  # Use 'y' instead of 'Y'
fitted_model = model_data_specification.fit()

# Hack to add data to legend
df['Data'] = 'Data'

# Scatter plot with OLS trendline
fig = px.scatter(df, x='x',  y='y', color='Data', trendline='ols', title='y vs. x')

# Add manual trendline (OLS fitted values)
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="OLS trendline")

# Add theoretical line
x_range_vals = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range_vals
fig.add_scatter(x=x_range_vals, y=y_line, mode='lines',
                name=str(beta0) + ' + ' + str(beta1) + ' * x',
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)

# Add horizontal line at y-bar (mean of y)
fig.add_scatter(x=x_range_vals, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer='png')
````

As observed in the visualization above, the ols trendline differs from the hypothetical line because it is calculated finding the least distance to the squared residuals (it uses 'squares' because it contemplates the distance from either side of the line, i.e. below or above the line). This ensures that both negative and positive data 'distance' is accounted for (observed data above and below the predicted line).

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>

### 10. Let's now consider just the (*n=160*) long wait times (as specified in the code below), and write code to do the following:

<br>

````python
import plotly.express as px
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.formula.api as smf

# Old Faithful dataset and defining long wait condition
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Original data with long waits
long_wait_data = old_faithful[long_wait]

# 1. Bootstrapping the slope coefficient from the actual data
n_bootstraps = 1000
bootstrapped_slopes = []

for _ in range(n_bootstraps):
    # Bootstrapped sample from the original data
    sample_data = long_wait_data.sample(n=long_wait.sum(), replace=True)
    
    # Fit OLS model to the bootstrapped sample
    model = smf.ols('duration ~ waiting', data=sample_data).fit()
    
    # Extract and store the slope coefficient
    bootstrapped_slopes.append(model.params['waiting'])

# Convert bootstrapped slopes to a Pandas DataFrame for easy plotting
bootstrapped_df = pd.DataFrame({'slope': bootstrapped_slopes})

# 2. Simulating data under the null hypothesis (slope = 0)
n_simulations = 1000
simulated_slopes = []

for _ in range(n_simulations):
    # Simulate Y values with slope = 0 under the null hypothesis
    old_faithful_simulation = long_wait_data.copy()
    old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation.waiting + \
        stats.norm(loc=0, scale=0.37).rvs(size=long_wait.sum())
    
    # Fit OLS model to the simulated data
    model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    
    # Extract and store the slope coefficient
    simulated_slopes.append(model.params['waiting'])

# Convert simulated slopes to a Pandas DataFrame for easy plotting
simulated_df = pd.DataFrame({'slope': simulated_slopes})

# 3. Plotting the bootstrapped slope distribution and simulated (null) slope distribution
fig = go.Figure()

# Histogram for bootstrapped slopes
fig.add_trace(go.Histogram(
    x=bootstrapped_df['slope'], nbinsx=50, name="Bootstrapped Slopes",
    histnorm='probability', opacity=0.75, marker=dict(color='blue')
))

# Histogram for simulated slopes under null hypothesis
fig.add_trace(go.Histogram(
    x=simulated_df['slope'], nbinsx=50, name="Simulated Slopes (Null)",
    histnorm='probability', opacity=0.75, marker=dict(color='orange')
))

# Add titles and labels
fig.update_layout(
    title="Bootstrapped vs Simulated Slopes Distribution",
    xaxis_title="Slope Coefficients",
    yaxis_title="Probability",
    barmode='overlay',
    showlegend=True
)

fig.show(renderer='png')

# 4. Calculating the 95% confidence interval for the bootstrapped slope distribution
lower_bound = np.percentile(bootstrapped_slopes, 2.5)
upper_bound = np.percentile(bootstrapped_slopes, 97.5)
print(f"95% Confidence Interval for Bootstrapped Slopes: ({lower_bound:.4f}, {upper_bound:.4f})")

# 5. Calculating the p-value from the simulated slopes distribution (under null hypothesis)
# This p-value corresponds to how many simulated slopes are greater than or equal to the observed bootstrapped mean
observed_mean_slope = np.mean(bootstrapped_slopes)
p_value = np.mean(np.array(simulated_slopes) >= observed_mean_slope)
print(f"P-value: {p_value:.4f}")

smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1]
````

Running both the boostrapped samples and the simulations yield a similar result; in the case of the null hypothesis testing, we get a p-value under 0.05, meaning we can reject the null hypothesis. In the case of the boostrapped samples, the 95% confidence interval does not overlap 0, which also indicates we can reject the null hypothesis.

`smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1]` indicates similar values both in p-value and 95% confidence interval.

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels

</details>

````python
import pandas as pd
import numpy as np
import plotly.express as px
from scipy import stats

# The "Classic" Old Faithful Geyser dataset
import seaborn as sns
old_faithful = sns.load_dataset('geyser')

# Assuming 'kind' is already a column where waiting times are split into 'short' and 'long'
long_wait_limit = 71
old_faithful['kind'] = np.where(old_faithful['waiting'] > long_wait_limit, 'long', 'short')

# Observed mean difference between 'short' and 'long' wait times
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# ------------------- Permutation Test -------------------
n_permutations = 1000
perm_diffs = []

for _ in range(n_permutations):
    # Shuffle the 'kind' labels
    shuffled_df = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)
    # Calculate the mean difference for the shuffled data
    shuffled_diff = shuffled_df.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    perm_diffs.append(shuffled_diff)

# Calculate the p-value: proportion of permuted differences greater than or equal to the observed difference
p_value = np.mean(np.abs(perm_diffs) >= np.abs(observed_diff))

# ------------------- Two-Sample Bootstrapping -------------------
n_bootstrap = 1000
bootstrap_diffs = []

for _ in range(n_bootstrap):
    # Resample within each 'kind' (short, long) with replacement
    boot_df = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    # Calculate the mean difference for the bootstrapped data
    boot_diff = boot_df.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrap_diffs.append(boot_diff)

# 95% confidence interval for the bootstrapped mean differences
conf_interval = np.percentile(bootstrap_diffs, [2.5, 97.5])

# ------------------- Plot Results -------------------
# Histogram for permutation test
fig_perm = px.histogram(perm_diffs, nbins=30, title="Permutation Test: Distribution of Mean Differences",
                        labels={'value': 'Mean Difference (Shuffled)'},
                        marginal='rug')
fig_perm.add_vline(x=observed_diff, line_dash="dash", line_color="red", annotation_text="Observed Diff", annotation_position="top right")
fig_perm.show(renderer='png')

# Histogram for bootstrap sampling distribution
fig_boot = px.histogram(bootstrap_diffs, nbins=30, title="Bootstrap: Distribution of Mean Differences",
                        labels={'value': 'Mean Difference (Bootstrap)'},
                        marginal='rug')
fig_boot.add_vline(x=observed_diff, line_dash="dash", line_color="red", annotation_text="Observed Diff", annotation_position="top right")
fig_boot.add_vrect(x0=conf_interval[0], x1=conf_interval[1], fillcolor="green", opacity=0.25, line_width=0, annotation_text="95% CI")
fig_boot.show(renderer='png')

# Output results
print(f"Observed Mean Difference: {observed_diff}")
print(f"P-value from Permutation Test: {p_value}")
print(f"95% Confidence Interval from Bootstrapping: {conf_interval}")
````

Quite obviously, both methods produce the same answer: we can reject the null hypothesis.

The first method suffles the labels and then calculates the difference in mean (or mean difference) many times (in this case, 1000). Then, in order to get the p-value, we have the observed mean difference from the unshuffled long and short wait times. This method also builds a mean difference distribution.

The second method creates a bootstrapped sample from each of the two kinds of waiting times and then gets the mean difference from the two (a 1000 times). The confidence interval is built on the mean difference sampling distribution obtained from the 'double bootstrapping' (and subsequent gathering of the mean difference). This is what is called 'double bootstrapping', as both samples are bootstrapped at the same time in order to get the mean difference.

#### Indicator variable?

The three methods all give the same answer, but the first method works directly with the simple linear regression model, simulating $Y_i$ values using the previously specified model parameters and the normally distributed error terms.

Perhaps a common trait of these methods is that a lot of the data points are simulated or resampled (in the case of bootstrapping). But the latter two methods focus on each separate 'kind' of waiting and then compare results with the observed difference (of the slope means, of course). The latter two methods are also dealing with the data points in the dataset (in order to bootstrap or to shuffle labels), as opposed to the first method, which is simulating data using the simple linear regression model with the indicator variable.
