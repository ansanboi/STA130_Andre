2. Mean: The average of the values in the column.
3. Std (Standard Deviation): This value represents the tendency of dispersion of values from the mean. This number provides insight into the consistency of the data.
4. Min (Minimum): Lowest value in a given column.
5. Percentiles (25th, 50th, 75th): Provides the boundaries for Q1, the median and the third quartile (Q3). These numbers indicate the spread of the data and help identify outliers. 25th percentile is the number under which 25% of the data falls. The 50th percentile is the number under/over which 50% of the data falls (the midpoint value). The 75th percentile is the number under which 75% of the data falls.
6. Max (Maximum): Highest value in a given column.

# .dropna() vs del

**`.dropna()`:** This method produces a new dataframe without columns or rows that have missing data.

**`del` function:** This function permanently deletes a column from a dataframe.

`.dropna()` allows us to experiment with our dataframe without the risk of permanently altering our data. It also provides the flexibility of eiher getting rid of columns *and* rows.

`del`, on the other and, deletes a column permanently. It is an efficient way of getting rid of data that we know we will not need. It is also useful to free up memory.

So, using `.dropna()` is better when keeping the data in its original state is important, and when we want to manipulate both columns and rows, while `del` is better-suited to handle cases where we need to free up space and we're certain that some data is not required. `del` is also useful when we just want to focus on some columns of the dataset.

# Using `del` and `.dropna()` together

Using `del` before using the `.dropna()` could be useful when dealing with columns you don't need. For example, the `'adult_male'` column could be considered irrelevant when we have another column making the distinction between man and woman passengers. This way, using `.dropna()` could take care of rows with missing data while `del` got rid of a redundant column. 

Another important thing to consider is that calling `.dropna()` to clean rows means that no column would disappear after; no columns would have missing data. Using `del` before would guarantee getting rid of a specific column beforehand.

````python
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)
df_rows = df.shape[0]
df_columns = df.shape[1]
print('number of rows:' , df_rows)
print('number of columns:' , df_columns)
print(df.columns)
print()
print()

df_copy = df.copy()
del df['adult_male']
print(df.columns)
print()

df_clean = df.dropna()
print(df_clean)

print()
print()
````

### `.groupby('col1')['col2'].describe()`

What this function does essentially is group the values in column 2 according to column 1. That is, for all the same values in column 1, the function shows the values in column 2 that correspond to them. Chaining `.describe()` gives the statistical summary of the column 2 values corresponding to the groups created from the data in column 1.

In my example above, I used the function to show me the statistical summary of the fares female and male passengers paid in the titanic. In this example, column 1 is 'sex' and column 2 is 'fare'. This shows me the summary of the fare vaules according to whether a passenger was female or male.

### 1. Difference Between `.dropna()` and `del`

- **`.dropna()`**:
  - Used in pandas to handle missing data by removing rows or columns that contain `NaN` values.
  - Returns a new DataFrame unless `inplace=True` is specified.
  - Customizable using parameters like `axis=0` (for rows) or `axis=1` (for columns).

- **`del`**:
  - A Python function used to delete entire columns or variables.
  - Alters the original DataFrame in place by removing the specified column.
  - Useful for freeing up memory, especially when working with large datasets.
